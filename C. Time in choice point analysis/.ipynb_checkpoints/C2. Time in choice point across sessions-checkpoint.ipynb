```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

from scipy.stats import wilcoxon
from scipy.stats import mannwhitneyu
%matplotlib inline
%run "C:\Users\anasofiaccruz\Desktop\cingulate_silencing_repo\general_purpose_functions.ipynb"
```

#### <font color='darkorange'> Read the data and remove baseline trials

```python
file_path = "C:/Users/anasofiaccruz/Desktop/Optogenetics processed data/no_outliers_1.5iqr_no_baseline.csv"
runs=pd.read_csv(file_path)
```

```python
counts=runs.groupby(['group','stim_condition']).count()['session']
counts
```

#### <font color='darkorange'> Boxplot time in choice point across sessions

```python
sns.set(style='white', context='talk')
plt.figure(dpi=300)
g = sns.catplot(kind='box', data=runs, col='group', x='session_nr', row='stim_condition',
                y='time_in_cp', hue='stim_condition', palette=['darkgray', 'royalblue', 'orangered'],
                height=4, aspect=2, saturation=.7) #linewidth=1, fliersize=1, legend=True)
g.set(ylim=(0,6),xlabel='session', ylabel='Time in choice point (s)', xticks=range(0,16,2),
      xticklabels=range(0,16,2))

axes=g.axes.flatten()
axes[0].set_title('NpHR-')
axes[1].set_title('NpHR+')
for i in range(2,6):
    axes[i].set_title('')
sns.despine()
```

#### <font color='darkorange'> Calculate the IQR values

```python
iqr = runs.groupby(['group', 'stim_condition', 'session_nr'])['time_in_cp'].describe().reset_index()
```

#### <font color='darkorange'> Plot IQR and median across sessions

```python
plt.figure(dpi=300)
sns.set(style='white', context='talk')
g=sns.relplot(kind='line',data=iqr, col='group', row='stim_condition', hue='stim_condition',
               y='50%', x='session_nr', ci=None, height=4, aspect=1.5,
              palette=['black', 'royalblue', 'orangered'], legend=False)

x=range(1,16)
g.set(xlim=[1,15], ylim=(0.5,1.5), xlabel='session', ylabel='I.Q.R')
colors = {0:'black', 1:'royalblue', 3:'orangered'}

for group, ax in g.axes_dict.items():
    group_iqr=iqr[(iqr['group']==group[1])&(iqr['stim_condition']==group[0])]
    color = colors[group[0]]
    ax.fill_between(x, group_iqr['25%'], group_iqr['75%'], alpha=.4, color=color)

ax = g.axes.flatten()  
ax[0].set(title='NpHR-')
ax[1].set(title='NpHR+')
for i in range(2,6):
    ax[i].set(title='')
```

#### <font color='darkorange'> Compare first and last session medians

```python
iqr_rats = runs.groupby(['group', 'rat','stim_condition', 'session_nr'])['time_in_cp'].describe().reset_index()
```

```python
first_last = iqr_rats[(iqr_rats['session_nr']==1) | (iqr_rats['session_nr']==15)]
first_last['session_nr'].unique()
first_last.to_csv('C:/Users/anasofiaccruz/Desktop/Optogenetics processed data/first_last_time_in_cp.csv')
```

```python
plt.figure(dpi=300)
sns.set(context='talk', style='white')
g=sns.catplot(kind='box',data=first_last, col='group', x='session_nr', y='50%', 
              hue='stim_condition', palette=['darkgray', 'royalblue', 'orangered'],
              height=5, aspect=1.3)
g.set(ylim=[0.2,1.5], ylabel='Median Time in choice point', xlabel='session number', xticks=[0,1],
     xticklabels=['1','15'])
ax=g.axes.flatten()
ax[0].set_title('NpHR-')
ax[1].set_title('NpHR+')
sns.despine()
```
