```python
from scipy.stats import ttest_1samp, wilcoxon

conditions = [0,1,3]
groups = ['CTRL', 'NPHR']
slope_inference_test=pd.DataFrame()
i=0
for c in conditions:
    for group in groups:
        df = slopes[(slopes['group']==group) & (slopes['stim_condition']==c)]
        t, p = ttest_1samp(df['slopes'], popmean=0)
        slope_inference_test.loc[i, 'group']=group
        slope_inference_test.loc[i,'condition']=c
        slope_inference_test.loc[i, 't']= t
        slope_inference_test.loc[i, 'p']=p
        i+=1
slope_inference_test  
```

### <font color='darkorange'> Cumulative errors across time

```python
#Add a session number to each run
all_rat_run_list = []
for rat in below15['rat'].unique():
    rat_runs = below15.loc[below15['rat']==rat]
    rat_runs.loc[:, 'session_nr'] = rat_runs.groupby(['session']).ngroup()
    all_rat_run_list.append(rat_runs)

runs = pd.concat(all_rat_run_list)
runs['session_nr']+=1
```

```python
#Keep the error runs only
errors = runs[runs['outcome']==0]
```

```python
sns.set(style='white', context='talk')
plt.figure(dpi=300)
none_errors=errors[errors['stim_condition']==0]
g = sns.displot(kind='ecdf', data=errors, col='group', x='session_nr', 
                hue='stim_condition', palette=['black', 'royalblue', 'orangered'],
                height=5, aspect=1,stat='proportion', linewidth=2)

ax = g.axes.flatten()
ax[0].set(title='NpHR-', xlabel='Session')
ax[0].axhline(0.5, linestyle='dashed')
ax[1].set(title='NpHR+', xlabel='Session')
ax[1].axhline(0.5, linestyle='dashed')
sns.despine()
```
