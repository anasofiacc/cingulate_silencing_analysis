## <font color='darkorange'> Read the data (test runs below 15 seconds)

```python
file_path = "E:\POSITION DATA\PRE PROCESSED/below15_runs.csv"
below15 = pd.read_csv(file_path)
```

### Remove baseline trials

```python
below15 = below15[(below15['stim_condition']!=-1)]
below15['stim_condition'].unique()
```

### Calculation of time spent in choice point

```python
below15['time_in_cp'] = below15['latency_to_cp_exit']-below15['latency_to_cp_entry']
# Drop runs with time in cp below zero (only 6 runs, from the same rat and session)
below15 = below15[below15['time_in_cp']>=0]
below15.groupby(['group','stim_condition'])['time_in_cp'].count()
```

### Scatter plot time in cp vs latency to cp entry
##### Because we remove runs with a latency to cp exit below 15 seconds, it is expected that runs with a higher value of latency to cp entry must low time in cp.

```python
plt.figure(dpi=300)
g = sns.relplot(
    kind='scatter', 
    data=below15,
    col='stim_condition',
    row='group',
    x='time_in_cp',
    y='latency_to_cp_entry')

ax = g.axes.flatten()
for group, ax in g.axes_dict.items():
    data = below15[(below15['group']==group[0])&
                   (below15['stim_condition']==group[1])
                  ]
    ax.axvline(np.median(data['time_in_cp']), color='black')
    ax.axhline(np.median(data['latency_to_cp_entry']), color='black')
    
    #ax.axvline(np.mean(data['time_in_cp']), color='green')
    #ax.axhline(np.mean(data['latency_to_cp_entry']), color='green')

sns.despine()
```

```python
file_path = "E:\POSITION DATA\PRE PROCESSED/test_runs.csv"
all_runs = pd.read_csv(file_path)
all_runs = all_runs[(all_runs['stim_condition']!=-1)]
all_runs['time_in_cp'] = all_runs['latency_to_cp_exit']-all_runs['latency_to_cp_entry']
# Drop runs with time in cp below zero (only 6 runs, from the same rat and session)
all_runs = all_runs[all_runs['time_in_cp']>=0]
#all_runs.groupby(['group','stim_condition'])['time_in_cp'].count()
plt.figure(dpi=300)
sns.relplot(
    kind='scatter', 
    data=all_runs,
    col='stim_condition',
    row='group',
    x='time_in_cp',
    y='latency_to_cp_entry')
plt.ylim([0,15])
sns.despine()
```
