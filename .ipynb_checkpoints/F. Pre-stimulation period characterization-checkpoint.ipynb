# Plot the learning curves of all rats in the experiment

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path='E:\learning_curves_all_rats.csv'
data=pd.read_csv(path, header=0)
data=pd.melt(data, id_vars=['SESSION'], var_name=['RAT'], value_name='PERF')
data['SESSION'].value_counts()
```

```python
sns.set(style="white", context='talk')
plt.figure(figsize=(5,4), dpi=300)

a = sns.lineplot(data=data, y='PERF', x='SESSION', color='slategray', ci='sd')
plt.plot(range(-10,0,1), [75]*10, '--', c='k')
a.set(ylim=[50,100], xlim=[-10,-1], ylabel='Performance(%)+/- st.d.', xlabel='sessions before stimulation protocol')
sns.despine()
```
