#  Does silencing the cingulate cortex during a test run leads to performance impairments? (timeouts removed)

#### In this notebook I want to determine whether silencing the cingulate cortex during test runs leads to a performance deficit.  

```python
# Import modules 
import pandas as pd
import numpy as np
import seaborn as sns
import os
import re
import matplotlib.pyplot as plt
from file_lists import get_file_list

from scipy.stats import shapiro, wilcoxon, mannwhitneyu

%matplotlib inline
%run data_prep_functions.ipynb
```

```python
file_path = "E:\POSITION DATA\PRE PROCESSED/below15_runs.csv"
below15 = pd.read_csv(file_path)
below15.head(3)
```

### <font color='royalblue'>Remove pre-illuminated trials


```python
below15 = below15[(below15['stim_condition']!=-1)]
below15['stim_condition'].unique()
```

### <font color='royalblue'> Calculate the performances for each rat, session and condition

### <font color='royalblue'> Inference testing

```python
sns.displot(pooled_perfs, x="performance", col="stim_condition", row="group",
            binwidth=2, height=3, facet_kws=dict(margin_titles=True))

normality_testing = {}
for group, data in pooled_perfs.groupby(['group', 'stim_condition'])['performance']:
    stat, p = shapiro(data)
    normality_testing[group] = (stat, p)
normality_testing = pd.Series(normality_testing)
normality_testing
```

#### The shapiro-Wilk test has confirmed that all but one distribution (NpHR+ , illumination condition) could have been withdrawn from a normal distribution (checked visually above too). Because for one of the distributions the null hypothesis was rejected, I performed two transformations in the four data distributions from pooled performance - square-root and log(e)/(10). I found that these transformations did not transform the NPHR+/illumination distribution into a distribution of data points that could have been withdrawn from a normal distribution, which was confirmed by the shapiro-Wilk test. I will resort to a non-parametric test, namely the Wilcoxon signed rank test, to compare the performance of subjects within each group, across the two experimental conditions. (two-tailed)

```python
# NpHR - group
ctrl_no = pooled_perfs[(pooled_perfs['group']=='CTRL') & (pooled_perfs['stim_condition']==0)]
ctrl_sample = pooled_perfs[(pooled_perfs['group']=='CTRL') & (pooled_perfs['stim_condition']==1)] 
ctrl_test = pooled_perfs[(pooled_perfs['group']=='CTRL') & (pooled_perfs['stim_condition']==3)]
```

```python
wilcoxon(ctrl_no['performance'], ctrl_test['performance'])
```

```python
# Remove CTRL2 No illumination performance
ctrl_no = ctrl_no[ctrl_no['rat']!='CTRL2']
```

```python
wilcoxon(ctrl_no['performance'], ctrl_sample['performance'])
```

```python
# NpHR+ group
nphr_no = pooled_perfs[(pooled_perfs['group']=='NPHR') & (pooled_perfs['stim_condition']==0)]
nphr_sample = pooled_perfs[(pooled_perfs['group']=='NPHR') & (pooled_perfs['stim_condition']==1)]
nphr_test = pooled_perfs[(pooled_perfs['group']=='NPHR') & (pooled_perfs['stim_condition']==3)]
```

```python
wilcoxon(nphr_no['performance'], nphr_sample['performance'], alternative='two-sided')
```

```python
wilcoxon(nphr_no['performance'], nphr_test['performance'], alternative='two-sided')
```

```python
from sklearn.linear_model import LinearRegression

def get_linear_regression_slopes(df, targetcol):
    
    lm = LinearRegression().fit(df[['session_nr']], df[[targetcol]])
    #slopes['slope']=lm.coef_[0]
    return lm.coef_[0][0]


slopes=pd.DataFrame()
slopes = session_perfs.groupby(['group', 'rat', 'stim_condition']).apply(get_linear_regression_slopes, 'performance')

slopes = slopes.reset_index().rename(columns={0:'slopes'})
```

```python
g = sns.stripplot(data=slopes, y='slopes', x='stim_condition',hue='group', s=12, 
             palette=['lightseagreen', 'orange'], alpha=.2, dodge=True)

sns.pointplot(data=slopes, y='slopes', x='stim_condition', hue='group', 
              dodge=.43, estimator=np.mean, join=False, ci='sd', color='blue',
              errwidth=2, palette=['lightseagreen', 'orange'])

g.legend_.remove()
plt.axhline(0, color='black', linestyle='dotted')
sns.despine()
```

```python
normality_testing = {}
for group, data in slopes.groupby(['group', 'stim_condition'])['slopes']:
    stat, p = shapiro(data)
    normality_testing[group] = (stat, p)

normality_testing = pd.Series(normality_testing)
normality_testing
```

```python
from scipy.stats import ttest_1samp, wilcoxon

conditions = [0,1,3]
groups = ['CTRL', 'NPHR']

slope_inference_test=pd.DataFrame()

i=0
for c in conditions:
    
    for group in groups:
        
        df = slopes[(slopes['group']==group) & (slopes['stim_condition']==c)]
        t, p = ttest_1samp(df['slopes'], popmean=0)

        slope_inference_test.loc[i, 'group']=group
        slope_inference_test.loc[i,'condition']=c
        slope_inference_test.loc[i, 't']= t
        slope_inference_test.loc[i, 'p']=p

        i+=1
slope_inference_test  
```

```python
def calc_session_performance_diffs(df):

    df_none = df[df['stim_condition']==0].sort_values(by='session_nr').reset_index(drop=True)
    df_sample = df[df['stim_condition']==1].sort_values(by='session_nr').reset_index(drop=True)
    df_test = df[df['stim_condition']==3].sort_values(by='session_nr').reset_index(drop=True) 
    
    df_test['test_none_diff']=df_test['performance']-df_none['performance']
    
    if df['rat'].iloc[0]=='CTRL2':
        df_test['sample_none_diff']=np.nan
    else:
        df_test['sample_none_diff']=df_sample['performance']-df_none['performance']
    
    final_df = df_test.drop(columns=['performance', 'stim_condition'])
    return final_df

performance_diff = session_perfs.groupby(['group','rat']).apply(calc_session_performance_diffs).reset_index(drop=True)
performance_diff.head()
```

```python
plt.figure(dpi=300)
sns.set(style="white", context='talk')

g = sns.relplot(
    data=performance_diff, 
    kind='line', col='group', hue='rat',
    x='session_nr', y='test_none_diff',  err_style='bars',
    linewidth=1, col_order=['CTRL','NPHR'], alpha=.2,
    height=4, aspect=1.3, palette=['orangered']*17, legend=False
)

for group, ax in g.axes_dict.items():
    sns.lineplot(data=performance_diff[performance_diff['group']==group], 
                 x='session_nr', y='test_none_diff',
                 estimator=np.mean, ci='sd', err_style='bars', 
                 color='orangered', linewidth=2, ax=ax)
    
    ax.axhline(linewidth=1, color='black', linestyle='dashed')

g.set(xlabel='Session', ylabel='Performance variation (%)+/-sd', 
      xticks=range(1,16,2), xticklabels=range(1,16,2),
     xlim=[0,16], ylim=[-70,50])
ax=g.axes.flatten()
ax[0].set(title='NpHR-')
ax[1].set(title='NpHR+')
sns.despine()
```

```python
def shapiro_normality_testing_for_group(df, grouping_cols_list, var):
    normality_testing = {}
    for group, data in df.groupby(grouping_cols_list)[var]:
        stat, p = shapiro(data)
        normality_testing[group] = (stat, p)

    normality_testing = pd.Series(normality_testing)
    return normality_testing

results = shapiro_normality_testing_for_group(test_diff_slopes, ['group'], 'slopes')
results
```
